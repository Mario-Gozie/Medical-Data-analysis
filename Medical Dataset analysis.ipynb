```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df1= df['smoker'] # what I did here, I created a dictionary to count th number of values in the smoker column of the df dataframe
countss = {} # so what the loop does is foe every value in the smoker column, df1.
# it searches for it in the countss dictionary. if its not there, it makes it a key and if its there, it adds one to it
# it them prints it as a key value pair

for entry in df1:
    if entry in countss.keys():
        countss[entry]+=1
    else:
        countss[entry]=1
print(countss)
```

```python
df.age.corr(df.bmi)
```

```python
dfmale = df[df['sex']== 'male']
dffemale = df[df['sex']== 'female']
```

```python
fig, ax = plt.subplots()
ax.scatter(dfmale['age'],dfmale['bmi'], c = 'b', label = 'male')
ax.scatter(dffemale['age'],dffemale['bmi'], c = 'r', label = 'female')
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Bmi')
plt.title('Bmi vs Age')
```

From what we can see here, there is no obvious relationship between Age and correlation as our correlation coefficient has confirmed earlier
